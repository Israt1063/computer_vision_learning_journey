```python
import pandas as pd
import numpy as np
from skimage.feature import graycomatrix, graycoprops
from PIL import Image

# Define GLCM feature extraction function
def extract_glcm_features(image_path):
    try:
        # Load image and convert to grayscale
        image = Image.open(image_path).convert('L')
        image_array = np.array(image)

        # Compute GLCM and features
        glcm = graycomatrix(image_array, distances=[1], angles=[np.pi/4], symmetric=True, normed=True)
        features = {
            "Image": image_path,
            "Dissimilarity": graycoprops(glcm, 'dissimilarity')[0, 0],
            "Correlation": graycoprops(glcm, 'correlation')[0, 0],
            "Homogeneity": graycoprops(glcm, 'homogeneity')[0, 0],
            "Contrast": graycoprops(glcm, 'contrast')[0, 0],
            "Energy": graycoprops(glcm, 'energy')[0, 0]
        }
        return features
    except Exception as e:
        print(f"Error processing {image_path}: {e}")
        return None

# Paths of images
image1_path = 'photo_2024-11-29_18-01-45.jpg'
image2_path = 'X-ray (1).jpeg'

# Extract features for both images
data = []
for image_path in [image1_path, image2_path]:
    features = extract_glcm_features(image_path)
    if features:
        data.append(features)

# Save features to CSV
df = pd.DataFrame(data)
csv_path = 'glcm_features.csv'
df.to_csv(csv_path, index=False)
print(f"Features saved to {csv_path}")

# Display head and tail of the CSV
print("\nHead of CSV:")
print(df.head())
print("\nTail of CSV:")
print(df.tail())
```

```python
import pandas as pd
import numpy as np
from skimage.feature import hog
from PIL import Image

def extract_hog_features(image_path):
    try:
        # Load and convert image to grayscale
        image = Image.open(image_path).convert('L')
        image = image.resize((128, 128))  # Resize image to standard dimensions
        image_array = np.array(image)

        # Compute HOG features
        features = hog(image_array, orientations=8, pixels_per_cell=(16, 16),
                       cells_per_block=(1, 1), visualize=False)
        return features
    except Exception as e:
        print(f"Error processing {image_path}: {e}")
        return None

# Paths for your images
image_paths = ['photo_2024-11-29_18-01-45.jpg', 'X-ray (1).jpeg']

# Extract HOG features for specific images
hog_data = []
for image_path in image_paths:
    features = extract_hog_features(image_path)
    if features is not None:
        hog_data.append([image_path, *features])

# Check if HOG data is available
if len(hog_data) > 0:
    # Dynamically create column names based on the number of features
    num_features = len(hog_data[0]) - 1  # Subtract 1 for the image column
    columns = ['Image'] + [f'HOG_Feature_{i}' for i in range(num_features)]

    # Save to CSV
    hog_df = pd.DataFrame(hog_data, columns=columns)
    csv_path = "hog_features.csv"
    hog_df.to_csv(csv_path, index=False)
    print(f"HOG features saved to {csv_path}")

    # Display head and tail of the CSV
    print("\nHead of CSV:")
    print(hog_df.head())
    print("\nTail of CSV:")
    print(hog_df.tail())
else:
    print("No HOG features were extracted. Please check your images and code.")
```
